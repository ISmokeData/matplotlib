```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
plt.boxplot(x)
plt.grid()
plt.show()
```

**Change shape**

```python
plt.boxplot(x, notch=True)
plt.show()
```

**Horizontal boxplot**

```python
plt.boxplot(x, vert=False)
plt.show()
```

**Width**

```python
plt.boxplot(x, widths=0.3)
plt.show()
```

**label**

```python
plt.boxplot(x, labels=["python"])
plt.show()
```

**Box color**

```python
plt.boxplot(x, patch_artist=True)
plt.show()
```

**show mean**

```python
plt.boxplot(x, showmeans=True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]
plt.boxplot(x)
plt.show()
```

**Join Outlier**

```python
plt.boxplot(x, whis=2)
plt.show()
```
