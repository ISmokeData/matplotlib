```python
import matplotlib.pyplot as plt
```

```python
ex = [0.1, 0.0, 0.0, 0.0]
y = ["c", "c++", "java", "python"]
x = [10, 20, 30, 40]
plt.pie(x, labels= y, explode= ex, autopct="%0.1f%%", shadow= True, radius=1.3, labeldistance=1.1, startangle=0, textprops = {"fontsize":15} )
plt.show()
```

**rotate**

```python
ex = [0.1, 0.0, 0.0, 0.0]
y = ["c", "c++", "java", "python"]
x = [10, 20, 30, 40]
plt.pie(x, labels= y, explode= ex, autopct="%0.1f%%", shadow= True, radius=1.3, labeldistance=1.1, startangle=0, textprops = {"fontsize":15}, counterclock=False )
plt.show()
```
