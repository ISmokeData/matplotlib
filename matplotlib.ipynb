```python
x = np.linspace(0,10, 200)
y = np.sin(x)
```

```python
plt.plot(x,y,'--r')
```

```python
plt.plot(x,y,'--b')
```

figure size control

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--b')
plt.show()
```

grid

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--b')
plt.grid()
plt.show()
```

custom color

```python
x = np.random.rand(50)
y = np.random.rand(50)
color = np.random.rand(50)
sizea = 1000*np.random.rand(50)
plt.scatter(x,y, c='#EDA550', s = sizea, alpha=.5)
plt.show()
```

3d projection

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.show()
```
