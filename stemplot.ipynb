**Stem Plot**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 3, 6, 7, 4, 3]

plt.stem(x,y)
plt.show()
```

**Line change**

```python
plt.stem(x,y, linefmt=":")
plt.show()
```

**change marker symbol**

```python
plt.stem(x,y, linefmt=":", markerfmt="r+")
plt.show()
```

```python
plt.stem(x,y, linefmt=":", markerfmt="ro")
plt.show()
```

**change bottom line**

```python
plt.stem(x,y, linefmt=":", markerfmt="ro",bottom=3)
plt.show()
```
