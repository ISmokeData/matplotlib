```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1,40, 50)
```

```python
x
```

```python
plt.hist(x,edgecolor="r")
plt.show()
```

**Standard bar**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x, edgecolor = "r", bins=l)
plt.show()
```

**Range Parameter**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins="auto", range=(0,100) ,edgecolor = "r")
plt.show()
```

**Reverse Histogram**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,"auto" ,edgecolor = "r", cumulative=-1)
plt.show()
```

**Y axis changes**

```python
plt.hist(x,bins="auto" ,edgecolor = "r" ,bottom=10) # y axis start from 10
plt.show()
```

**Align**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins=l,edgecolor = "r", align="mid")
plt.show()
```

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins=l,edgecolor = "r", align="mid")
plt.show()
```

**Type**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins=l,edgecolor = "r", histtype= "step")
plt.show()
```

**orientation**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins=l,edgecolor = "r",orientation="horizontal")
plt.show()
```

**width**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins=l,edgecolor = "r", rwidth =0.5)
plt.show()
```

**axvline**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.hist(x,bins=l,edgecolor = "r", rwidth =0.5)
plt.axvline(20, color="r", label="line")
plt.legend()
plt.show()
```

**grid**

```python
l = [0, 5, 10, 15, 20, 25, 30, 35, 40]
plt.figure(figsize =(10,6))
plt.hist(x,bins=l,edgecolor = "r", rwidth =0.5)
plt.axvline(20, color="r", label="line")
plt.legend()
plt.grid()
plt.show()
```
